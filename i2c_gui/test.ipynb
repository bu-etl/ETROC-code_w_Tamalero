```python
from i2c_gui.chips import ETROC2_Chip


chip = ETROC2_Chip()


broadcast_handle,_,_ = chip.get_indexer("broadcast")
row_handle,_,_ = chip.get_indexer("row")
col_handle,_,_ = chip.get_indexer("column")

IBSel_handle = chip.get_indexed_var("ETROC2", "Pixel Config", "IBSel")

chip.read_all_address_space("ETROC2")

IBSel_handle.set("0x7")
broadcast_handle.set(True)
row_handle.set('0xf')
col_handle.set('0xf')

chip.write_all_address_space("ETROC2", write_check=False)
```

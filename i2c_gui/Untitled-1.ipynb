```python
import logging
import i2c_gui2
import time
```

```python
port = "/dev/ttyACM0"
clock = 100

conn = i2c_gui2.USB_ISS_Helper(port, clock, dummy_connect = False)

print(f"USB-ISS Firmware Version: {conn.fw_version}")
print(f"USB-ISS Serial Number: {conn.serial}")

chip_logger = logging.getLogger("Chip")

etroc_address = 0x60
ws_address = 0x40
chip = i2c_gui2.ETROC2_Chip(etroc_address, ws_address, conn, chip_logger)

start_time = time.time()
# chip.read_all()
# chip.read_all_block("ETROC2", "Peripheral Config")
# chip.read_all_block("ETROC2", "Pixel Config", full_array=True)
chip.read_all_efficient()
end_time = time.time()
chip.save_config("test_dump.pckl")

print("--- %s seconds ---" % (end_time - start_time))
```
